```python
df1= df2.drop(['Commodity Name','Sub Variety', 'City Name','Type','Crop', 'Repack', 'Trans Mode','Mostly Low','Mostly High','Storage','Environment','Origin District','Grade', 'Unit of Sale', 'Item Size','Quality','Color', 'Condition', 'Appearance','Repack'], axis = 1)
df1
```

```python
price = (df1['Low Price'] + df1['High Price']) / 2

month = pd.DatetimeIndex(df1['Date']).month
```

```python
new_pumpkins = pd.DataFrame({'Month': month, 'Package': df1['Package'], 'Low Price': df1['Low Price'],'High Price': df1['High Price'], 'Price': price})
```

```python
new_pumpkins.drop(['Low Price','High Price'], axis = 1)
```

```python
new_pumpkins['Package']
```

```python
import seaborn as sns; sns.set_theme()
```

```python
pumpkins = new_pumpkins[new_pumpkins['Package'].str.contains('bushel', case=True, regex=True)]
```

```python
pumpkins
```

```python
new_pumpkins.loc[new_pumpkins['Package'].str.contains('1 1/9'), 'Price'] = price/(1 + 1/9)

new_pumpkins.loc[new_pumpkins['Package'].str.contains('1/2'), 'Price'] = price/(1/2)
```

```python
import matplotlib.pyplot as plt
```

```python
price = new_pumpkins.Price
month = new_pumpkins.Month
plt.scatter(price, month)
plt.show()
```
